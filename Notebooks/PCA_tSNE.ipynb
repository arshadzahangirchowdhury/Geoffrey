```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
data = load_wine()
# print(data.feature_names)
# print(data.target_names)
# print(data.target)
X = data.data
y = data.target
model = PCA(n_components=2)
X_transformed = model.fit_transform(X)
# print(X_transformed.shape)
# model.components_
print(model.explained_variance_ratio_)
plot=plt.scatter(X_transformed[:,0], X_transformed[:,1], c=y)
plt.legend(handles=plot.legend_elements()[0], labels=data.target_names.tolist())

```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.manifold import TSNE
np.random.seed(42)
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X = np.concatenate([X_train, X_test]).reshape(-1, 28*28) / 255.0
y = np.concatenate([y_train, y_test])

print(X.shape)
print(y.shape)


# Sample 6000 images
sample_indices = np.random.choice(len(X), 6000, replace=False)
X_sample = X[sample_indices]
y_sample = y[sample_indices]

tsne = TSNE(n_components=2, perplexity=25, random_state=42)
X_transformed = tsne.fit_transform(X_sample)

plt.figure(figsize=(10, 8))
cmap = plt.cm.get_cmap('tab10', 10)

for i in range(10):
    indices = y_sample == i
    plt.scatter(X_transformed[indices, 0], X_transformed[indices, 1], c=[cmap(i)], label=i, alpha=0.5, s=50)

plt.legend(title="Digits", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.title("t-SNE with 2 components and perplexity 25 on 6000 MNIST digits")
plt.show()
```
